```python
import numpy as np
import pandas as pd
import sklearn.linear_model as skl_lm
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('logger.txt', sep="  ", header=None)
data.columns = ['n','t']
data.plot(x='n',y='t',logy=True, logx=True)
x = range(0,2000000)
y = (x*np.log(x))/6000000
plt.loglog(x,y,'-')
plt.grid(True,which="both",ls="-")
plt.legend(['Quicksort','nlog(n)'])
plt.xlabel('Array length')
plt.ylabel('Execution time [s]')
```
